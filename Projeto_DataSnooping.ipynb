#Data Snooping + GNSS

O Data Snooping Iterativo (Iterative Data Snooping - IDS) é o procedimento mais conhecido para identificação de outliers em dados geodésicos.

Para o caso de observações descorrelacionadas (sem covariâncias), sendo $m$ a qtd de observações, $v_i$ o resíduo e $\sigma_{v_i}$ o desvio padrão do mesmo, a estatística de teste da $i-ésima$ observação é dada por:

$w_i=\frac{v_{i}}{\sigma_{v_i}}, \;i=1,2,...,m$

O máximo $w_i$ absoluto obtido ($máx |w_i|$) é então comparado com determinado valor crítico. Uma conduta ainda usual é obter o valor crítico $|Z_{\alpha/2}|$ diretamente na tabela estatística da distribuição normal, sendo $\alpha$ o nível de significância do teste (definido pelo usuário). Se $max |w_i| > |Z_{\alpha/2}|$, a respectiva observação é classificada como outlier e usualmente excluída da rede. O mesmo procedimento, iniciando pelo ajustamento pelo MMQ da “nova” rede (sem o outlier identificado), é repetido até que nenhuma observação seja identificada como outlier.
